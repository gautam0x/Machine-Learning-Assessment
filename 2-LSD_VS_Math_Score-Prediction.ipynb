## We Are Going To Predict : Is LSD Drug Good Or Bad For Math Exam

```python
import pandas as pd
```

```python
data = pd.read_csv('DATA/lsd_math_score_data.csv')
```

```python
print(data)
```

```python
type(data)
```

### PLOT GRAPH 

* Time_Delay_in_Minutes v/s LSD_ppm
* LSD_ppm v/s Avg_Math_Test_Score

```python
time = data[['Time_Delay_in_Minutes']]
LSD_ppm = data[['LSD_ppm']]
math_score =  data[['Avg_Math_Test_Score']]
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(time,LSD_ppm,color='#e74c3c', linewidth=3)
plt.title('Tissue concentration of LSD over time', fontsize=17)
plt.xlabel('Time in Minutes', fontsize=14)
plt.ylabel('Tissue LSD ppm', fontsize=14)
plt.text(x=0, y=-0.5, s='Wagner et al. (1968)', fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(1,2,2)
plt.plot(LSD_ppm,math_score,color='#77f43c', linewidth=3)
plt.title('Arithmetic vs LSD-25', fontsize=17)
plt.xlabel('Tissue LSD ppm', fontsize=14)
plt.ylabel('Performance Score', fontsize=14)

plt.show()
```

### Predict Score and LSD_PPM

* predict using linear regression and plot 

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()

regression.fit(LSD_ppm,math_score)
print('Theta1 : ', regression.coef_[0][0])
print('Intercept: ', regression.intercept_[0])
print('R-Square: ', regression.score(LSD_ppm, math_score))
predicted_score = regression.predict(LSD_ppm)
```

```python
plt.title('Arithmetic vs LSD-25', fontsize=17)
plt.xlabel('Tissue LSD ppm', fontsize=14)
plt.ylabel('Performance Score', fontsize=14)
plt.style.use('fivethirtyeight')

plt.scatter(LSD_ppm, math_score, color='blue', s=100, alpha=0.7)
plt.plot(LSD_ppm, predicted_score, color='red', linewidth=3)
plt.show()
```
