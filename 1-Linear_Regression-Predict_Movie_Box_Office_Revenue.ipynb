## Prediction Of How Much Movie Earn On Perticular Budget

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('DATA/cost_revenue.csv')
```

```python
data.describe()
```

```python
# Read x and y parameter from csv
x = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
# Predict the slope
regression = LinearRegression()
regression.fit(x,y)
```

## Using Slope Formula 
$y = mx + c$

or

$y = c + mx$

or
#### h<sub>&theta;</sub>(x) = &theta;<sub>o</sub> + &theta;<sub>1</sub>x

where &theta;<sub>1</sub> is coefficient and &theta;<sub>0</sub> is intercept

```python
regression.coef_ ####theta_1
```

```python
regression.intercept_ #####theta_0
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3) # plot the x,y data
plt.plot(x,regression.predict(x),color='red',linewidth=3) # predict the slope of x,y
plt.title('Movie Budget vs Global Revenue')
plt.xlabel('Movie Budget')
plt.ylabel('Global Revenue')
plt.ylim(0,)
plt.xlim(0,)
plt.show()
```

```python
regression.score(x,y) #0.549 maeans 54.9% or 55% variation in revenue 
```
